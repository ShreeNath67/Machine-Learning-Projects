# House Price Prediction using Linear Regression

## Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Define the Dataset

```python
# House sizes in 1000 sqft
X = np.array([6, 7, 8, 9, 10, 5, 6.5, 8.5, 7.2, 9.5,
              4.5, 5.5, 6.8, 7.8, 9.2, 10.5, 4, 7.5, 8.2, 6.2])

# Prices in ₹ lakhs
y = np.array([45, 50, 55, 60, 65, 40, 48, 58, 51, 63,
              38, 44, 52, 56, 62, 70, 36, 54, 57, 47])

# Number of data points
m = len(X)
```

## Step 3: Visualize the Dataset

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel("House Size (1000 sqft)")
plt.ylabel("Price (in ₹ lakhs)")
plt.title("House Price vs Size")
plt.grid(True)
plt.legend()
plt.show()
```

## Step 4: Initialize Parameters

```python
w = 0  # Weight (slope)
b = 0  # Bias (intercept)
alpha = 0.01  # Learning rate
iters = 1000  # Number of iterations
```

## Step 5: Define Cost Function

<img src="Cost Function.png" width="600">

```python
# Cost function - Mean Squared Error
def compute_cost(X, y, w, b):
    m = len(X)
    total_cost = 0
    for i in range(m):
        y_pred = w * X[i] + b
        total_cost += (y_pred - y[i]) ** 2
    return total_cost / (2 * m)
```

## Step 6: Define Gradient Descent Algorithm

<img src="Gradient Descent.png" width="600">

```python
# Gradient Descent Function is use to minimize the error
def gradient_descent(X, y, w, b, alpha, iters):
    m = len(X)
    cost_history = []

    for epoch in range(iters):
        dw = 0
        db = 0
        for i in range(m):
            y_pred = w * X[i] + b
            dw += (y_pred - y[i]) * X[i]
            db += (y_pred - y[i])
        dw /= m
        db /= m

        # Update parameters
        w -= alpha * dw
        b -= alpha * db

        # Store cost
        cost = compute_cost(X, y, w, b)
        cost_history.append(cost)

        if epoch % 100 == 0:
            print(f"iters {iter}: Cost={cost:.4f}, w={w:.4f}, b={b:.4f}")

    return w, b, cost_history
```

## Step 7: Train the Model

```python
# Run the training
w_opt, b_opt, costs = gradient_descent(X, y, w, b, alpha, iters)
```

## Step 8: Plot the Regression Line

```python
plt.figure(figsize=(8, 5))
y_pred = w_opt * X + b_opt
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel("House Size (1000 sqft)")
plt.ylabel("Price (in ₹ lakhs)")
plt.title("Optimized Linear Regression Fit")
plt.grid(True)
plt.legend()
plt.show()
```

## Step 9: Plot Cost Function Over iterations

```python
# Cost function reduction over time
plt.figure(figsize=(8, 5))
plt.plot(range(iters), costs, color='green')
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Cost Reduction over Training")
plt.grid(True)
plt.show()
```

## Step 10: Final Output

```python
print(f"Final Optimized Weight (w): {w_opt:.4f}")
print(f"Final Optimized Bias (b): {b_opt:.4f}")
```
